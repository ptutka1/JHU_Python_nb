```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD  #Stochastic Gradient Descent

import numpy as np
# fix random seed for reproducibility
np.random.seed(7)
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
n = 50
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

```python
plt.scatter(Hidden, Tanh_2_neurons, color='blue', label=None)
plt.scatter(Hidden, Tanh_3_neurons, color='orange', label=None)
plt.scatter(Hidden, Tanh_4_neurons, color='red', label=None)
plt.plot(Hidden, Tanh_2_neurons, color='blue', label='2 Neurons')
plt.plot(Hidden, Tanh_3_neurons, color='orange', label='3 Neurons')
plt.plot(Hidden, Tanh_4_neurons, color='red', label='4 Neurons')
plt.legend()
plt.xlim()
plt.ylim()
plt.title("Neural Network Hyperbolic Tangent Error (XOR) ")
plt.xlabel("Number of Hidden Layers")
plt.ylabel("Loss (Binary Entropy)")
```

```python
# create model
model = Sequential()
model.add(Dense(4, input_dim=2, activation='tanh'))

model.add(Dense(4, activation='tanh'))
model.add(Dense(4, activation='tanh'))
#model.add(Dense(3, activation='softmax'))
#model.add(Dense(2, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
# Fit the model
model.fit(X, y, epochs=400, batch_size=2)
# evaluate the model
scores = model.evaluate(X, y)
#print(model.predict_proba(X).reshape(4*n))
```

```python
mushroom= pd.read_csv('C:\JHU_Python\mushrooms.csv')
```

```python
target = 'class' # The class we want to predict
labels = mushroom[target]
features = mushroom.drop(target, axis=1) # Remove the target class from the dataset
```

```python
categorical = features.columns # Since every fearure is categorical we use features.columns
features = pd.concat([features, pd.get_dummies(features[categorical])], axis=1) # Convert every categorical feature with one hot encoding
features.drop(categorical, axis=1, inplace=True) # Drop the original feature, leave only the encoded ones

labels = pd.get_dummies(labels)['p'] # Encode the target class, 1 is deadly 0 is safe
```

```python
# create model
model = Sequential()
model.add(Dense(3, input_dim=117, activation='tanh'))

model.add(Dense(3, activation='tanh'))
#model.add(Dense(3, activation='tanh'))
#model.add(Dense(2, activation='tanh'))
#model.add(Dense(2, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
# Fit the model
model.fit(features, labels, epochs=100, batch_size=2)
# evaluate the model
scores = model.evaluate(features, labels)
#print(model.predict_proba(X).reshape(4*n))
```
