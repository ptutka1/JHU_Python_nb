# 1. Show the most common words in the article.

```python
from collections import Counter
from spacy.attrs import ORTH
```

```python
def cleanup(token, lower = True):
    if lower:
       token = token.lower()
    return token.strip()
```

```python
noisy_pos_tags = ['PROP']
min_token_length = 2 # excludes words such as "a" and other tokens "punct" from the model. 
```

```python
def isNoise(token):     
    is_noise = False
    if token.pos_ in noisy_pos_tags:
        is_noise = True 
    elif token.is_stop == True:
        is_noise = True
    elif len(token.string) <= min_token_length:
        is_noise = True
    return is_noise 
```

The above for loop only includes tokens that are words that are at least two or more characters and do not include punctuation. Since the article is about Austin and police activities both words are mentioned Austin 23 and police 21 times. 

```python
cleaned_list = [cleanup(word.string) for word in WP_processed if not isNoise(word)]
Counter(cleaned_list).most_common(10)
```

```python
counts = Counter()
for token in WP_processed:
    counts[token.text] += 1 # Equivalently, token.text
    c = counts.most_common(10)  
c
```

# 3. Find a subject/object relationship through the dependency parser in any sentence.

I analyzed the first line of the news article to find the "bomber" is the subject of the sentence and "Austin" is the direct object of the sentence. I think the model did a great job of breaking apart the sentence and analyzing the denpendencies. 

```python
Line1 = nlp2("""The suspected serial bomber who terrorized Austin with a string of exploding packages died early Wednesday after detonating an explosive device in his vehicle as police closed in on him, authorities said.""")
```

```python
def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)
```

```python
for sentence in Line1.sents:
    pr_tree(sentence.root, 0)
    print('-------------------------------------------')
```

# 5. Find Entites and their dependency (hint: entity.root.head)

```python
def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)
```

```python
for entity in Line1.ents:
    pr_tree(entity.root.head, 0)
    print('-------------------------------------------')
```
